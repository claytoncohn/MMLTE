```python
import pandas as pd

S18_PATH = "drive/MyDrive/Clayton/20230420_MMLTE/S18.csv"
S18 = pd.read_csv(S18_PATH)

S18.head()
```

## Extract NLP

```python
modalities = set(["PROS", "TRANS", "TEXT", "SPECT", "AFFECT"])

UUIDs = set()
for i, row in S18.iterrows():
    mods = row["Modalities"].split(",")
    for mod in mods:
        if mod in modalities:
            if mod == "AFFECT":
              if "AUDIO" in row["Data Collection Mediums"]:
                UUIDs.add(row["UUID"])
            else:
              UUIDs.add(row["UUID"])
```

```python
df_nlp = S18[S18["UUID"].isin(UUIDs)]
df_nlp.reset_index(drop=True, inplace=True)

print(len(df_nlp))
df_nlp.head(50)
```
